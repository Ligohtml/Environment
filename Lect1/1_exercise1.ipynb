```python
import numpy as np
```

```python
mean = 0
std = 1
dim = 1000
data = np.zeros(dim)
```

```python
g = np.random.normal(mean,std,dim)
```

```python
def mean(g,dim):
  sum=0
  for i in range(dim):
    sum = g[i] + sum
  return sum/dim

def stddev(g,dim,mean):
  sum_sq = 0
  for i in range(dim):
    sum_sq = (g[i]-mean)**2 + sum_sq
  return sum_sq/dim


mean_temp = mean(g, dim)
std_temp = stddev(g, dim,mean_temp)
print(f"Mean: {mean_temp}")
print(f"Standard Deviation: {std_temp}")
```

```python
import timeit

%timeit -n 10000 -r 3 mean(g, dim)
%timeit -n 10000 -r 3 np.mean(g)
```
